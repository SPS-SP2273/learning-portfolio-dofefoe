# 2 Comparing the two ‘dialects’

### pyplot Interface

```python
from matplotlib import pyplot as plt
import numpy as np
x = np.linspace(-np.pi, np.pi, num=100)
cos_x = np.cos(x)
sin_x = np.sin(x)
plt.plot(x, cos_x, label='cos x')
plt.plot(x, sin_x, label='sin x')
plt.legend()

```

### OO Interface

```python
fig, ax = plt.subplots(nrows=1, ncols=1)
ax.plot(x, cos_x, label='cos x')
ax.plot(x, sin_x, label='sin x')
ax.legend() #generates same plot as cell above
```

# 5 Multiple rows and columns

### Using ax

```python
fig, ax = plt.subplots(nrows=2, ncols=1) #figure split in 2 rows, displaying 2 subplots

ax[0].plot(x, cos_x, label='cos x')
ax[1].plot(x, sin_x, label='sin x')

ax[0].legend()
ax[1].legend()
```

```python
fig, ax = plt.subplots(nrows=2, ncols=1,
                       figsize=(5, 5),
                       sharex=True)
ax[0].plot(x, cos_x, label=r'$\cos(x)$')
ax[0].fill_between(x, 0, cos_x, alpha=.25)
ax[1].plot(x, sin_x, label=r'$\sin(x)$')
ax[1].fill_between(x, 0, sin_x, alpha=.25)

for a in ax:
    a.legend()
    a.grid(alpha=.25)
    a.set_ylabel('$y$', rotation=0) #note difference plt.ylabel() vs ax.set_ylabel()

ax[1].set_xlabel('$x$')

fig.suptitle(r'$\sin(x)$ and $\cos(x)$') #adds super title to figure
fig.tight_layout() #smaller distance between plots
```

## 5.1 Sharing axes

```python
fig, ax = plt.subplots(nrows=2, ncols=2,
                       figsize=(5, 5),
                       sharex='col', sharey='col')

# Some variables to access the axes and improve readabilty
top_left, top_right, bottom_left, bottom_right = ax.flatten() #now each new var contains just 1 axes

top_left.set_title('Top Left')
top_right.set_title('Top Right')
bottom_left.set_title('Bottom Left')
bottom_right.set_title('Bottom Right')
#alternatively
#ax[0, 0].set_title('Top Left')
#ax[0, 1].set_title('Top Right')
#ax[1, 0].set_title('Bottom Left')
#ax[1, 1].set_title('Bottom Right')

top_left.hlines(y=1, xmin=0, xmax=4)
top_right.hlines(y=2, xmin=0, xmax=5)
bottom_left.hlines(y=3, xmin=0, xmax=4)
bottom_right.hlines(y=4, xmin=0, xmax=5)

for a in ax.flatten():
    a.grid(alpha=.25)

plt.tight_layout()
```

## 5.1 Sharing axes

```python
fig, ax = plt.subplots(
    nrows=2, ncols=2,
    figsize=(5, 5), sharex = True, sharey = "col"
) #without sharex and sharey, each of the 4 subplots autoscales
#accepted arg for param sharex or sharey are: 
#"True" - makes all axes use sage range
#"col" - use same range for columns
#"row" - use same range for rows
ax[0, 0].set_title('Top Left')
ax[0, 1].set_title('Top Right')
ax[1, 0].set_title('Bottom Left')
ax[1, 1].set_title('Bottom Right')

ax[0, 0].hlines(y=1, xmin=0, xmax=4)
ax[0, 1].hlines(y=2, xmin=0, xmax=5)
ax[1, 0].hlines(y=3, xmin=0, xmax=4)
ax[1, 1].hlines(y=4, xmin=0, xmax=5)

for a in ax.flatten():
    a.grid(alpha=.25)

plt.tight_layout()
```

# 6 Other useful plots

## 6.1 Histograms

### A quick helper function

```python
def get_plot_data():
    data = {}
    filename = 'sg-gov-graduate-employment-survey_basic_monthly_mean_all.csv'
    data['All'] = np.loadtxt(filename, skiprows=1)

    filename = 'sg-gov-graduate-employment-survey_basic_monthly_mean_nus.csv'
    data['NUS'] = np.loadtxt(filename, skiprows=1)

    return data

```

### The histogram

```python
plt.style.use('bmh')
data = get_plot_data()

# bins specifies how many bins to split the data
plt.hist([data['All'], data['NUS']], bins=50, label=['All', 'NUS'])
plt.xlabel('Mean of Basic Montly Earning (S$)')
plt.ylabel('Number of Students')
plt.legend()
```

## 6.2 Scatter plots

```python
plt.style.use("seaborn-v0_8-darkgrid")

data = get_plot_data()

for label, numbers in data.items(): #loops once, once for NUS, once for ALl
    x = range(len(numbers)) #generates list for size of numbers (of category NUS or all)
    y = numbers
    plt.scatter(x, y, label=label, alpha=.5)

plt.xlabel('Position in the list')
plt.ylabel('Mean of Basic Montly Eraning (S$)')
plt.legend(loc = "best") #"best" loc switches between upper left and upper right for some reason
```

## 6.3 Bar charts

```python
student_numbers = {'Life Sciences': 14,
                   'Physics': 12,
                   'Chemistry': 8,
                   'Comp. Biology': 1} #dummy values
majors = student_numbers.keys()
numbers = student_numbers.values()

plt.style.use('ggplot')
plt.bar(majors, numbers) #barh() for horizontal bars
plt.xlabel('Majors')
plt.ylabel('Number of Students')
```

## 6.4 Pie charts

```python
student_numbers = {'Life Sciences': 14,
                   'Physics': 12,
                   'Chemistry': 8,
                   'Comp. Biology': 1}
majors = student_numbers.keys()
numbers = student_numbers.values()

plt.style.use('fivethirtyeight')
plt.pie(numbers, 
        labels=majors,
        autopct='%1.2f%%',   # How to format the percentages
        startangle=-90       #comp. Bio starts at 6 o clock position        
        )
plt.title('Percentage of each major')
```
