```python
from matplotlib import pyplot as plt
import numpy as np
# Some data for plotting
xb = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]
y_3 = [0, 4, 8, 12, 16, 20, 24]

x = np.linspace(0, 6, 100) # returns evenly spaced values within the 2 intervals
np_x = np.array(x) # changing to array makes it easier to do math
np_y_1 = np.array(y_1)
np_y_2 = np.array(y_2)

x2 = np_x**2                  
sin_x = np.sin(x)

x_error, y_error = [1,2,3,4,5,6,7],[7,6,5,4,3,2,1] 
#can be single value or list. However list has to be same number as plot points

# Let's start plotting
plt.plot(x, x2, label='Y values',
         color='red', linestyle='dashed')
plt.legend(loc = "lower left")#declared before next plot function call, so wld be split nxt legend call (if any)
plt.twinx() #next plot would have its own y axis
plt.plot(x, sin_x,
         label='Y2 values', color='blue') #second line in same graph
plt.errorbar(xb, y_3,
             xerr=x_error, yerr=y_error,
             label="Y3 with errors",
             color="green")
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(c = "blue", lw = "3",ls ="dotted",alpha= 0.25) #abbrevations for color, linewidth, linestyle
#param inc visible(bool), which (major/minor/both), axis (x/y/both) to emphasis which lines are being altered
plt.legend(loc='best') #accepted loc are "upper/centre/lower left/centre/right" eg "top left" "centre"
#also has "best", which looks for least overlap among the 9 locations
plt.savefig('simple-plot.png', dpi=150) #need to call before show. data format will be speciialised
plt.show() #displays open figures, not necessary on notebook

```
