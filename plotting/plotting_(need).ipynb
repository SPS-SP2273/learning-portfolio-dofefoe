```python
from matplotlib import pyplot as plt
import numpy as np
# Some data for plotting
xb = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]
y_3 = [0, 4, 8, 12, 16, 20, 24]

x = np.linspace(0, 6, 100) # returns (100) evenly spaced values within the 2 intervals (0 to 5)
np_x = np.array(x) # changing to array makes it easier to do math
np_y_1 = np.array(y_1)
np_y_2 = np.array(y_2)

x2 = np_x**2                  
sin_x = np.sin(x)

plt1 = plt.plot(x, x2, label='Y values',
         color='red', linestyle='dashed')
plt.legend(loc = "lower left")#declared before next plot function call, so wld be split nxt legend call (if any), take note of overlap
plt.twinx() #next plot would have its own y axis but in same plot
plt2 = plt.plot(x, sin_x,
         label='Y2 values', color='blue') #second line graph in same figure

plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(c = "blue", lw = "3",ls ="dotted",alpha= 0.25) #abbrevations for color, linewidth, linestyle
#param inc visible(bool), which (major/minor/both), axis (x/y/both) to emphasis which lines are being altered
plt.legend(loc='best') #accepted loc are "upper/centre/lower left/centre/right" eg "top left" "centre"
#also has "best", which looks for least overlap among the 9 locations
plt.savefig('simple-plot.png', dpi=150) #data format can be specified, pdf, jpeg, png etc
plt.show() #displays open figures, not necessary on notebook

#ax_overall = ax1 + ax2
#labs = [l.get_label() for l in plt_overall]
#ax.legend(plt_overall, labs, loc=0)
#it seems axis is needed to combine the 2 legends cuz of its nature as an array
#alternatively draw invisible line in one of the plot and label that with the other lines value
```

```python
# Some data for plotting
x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]
y_3 = [0, 4, 8, 12, 16, 20, 24]
#x_error, y_error = .1, 0.75
x_error, y_error = [.1,3,.5,2,1,.6,.7], [0.75,1,2,3,4,5,6] #error can be static value or a list of size equal to x and y
                                                            #only accepts positive floats

# Lets start plotting
plt.plot(x, y_1, label='Y1 values', color='red', linestyle='dashed',)
plt.plot(x, y_2, label='Y2 values', color='blue', )
plt.errorbar(x, y_3, xerr=x_error, yerr=y_error,
             label='Y3 with errors', color='green')
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(loc='upper left')
```

# 4 A real example: Global Warming

```python

data = np.loadtxt('Complete_TAVG_daily.txt', #fast text file reader from numpy
                   skiprows=24) #skip lines
#data stored as 2d array containing date list in first element and anomaly size list in last element
date = data[:, 0]
anomaly = data[:, -1]
[print(x) for x in anomaly if x <-2] #did I get different data from notes?
plt.style.use('fivethirtyeight') #needs to be declared before plot is drawn ie nxt line
plt.plot(date, anomaly, alpha=.5)

plt.ylim([-8, 8]) #y axis boundaries
plt.hlines(0, date[0], date[-1], linestyle='--', colors='grey') #horizontal line
plt.xlabel('Date')
plt.ylabel('Temperature Anomaly')
plt.title('Temperature anomaly\n(Relative to  average from Jan 1951 - Dec 1980.)')
#plt.xkcd() #xkcd-ifies the style

plt.show()
```
